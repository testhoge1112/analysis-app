```python
import os
import glob
import pandas as pd
from datetime import datetime, timedelta

# ===== 設定 =====
DATA_DIR   = r"C:\Users\akafu\analytics\aim_machi\data"            # 読み込み元
OUTPUT_DIR = r"C:\Users\akafu\analytics\aim_machi\notebook\result"  # 出力先

START_DATE = "2025-06-13"
END_DATE   = "2025-06-30"


def daterange(start_date: datetime, end_date: datetime):
    d = start_date
    while d <= end_date:
        yield d
        d += timedelta(days=1)


def find_file_for_date(d: datetime) -> str | None:
    """
    その日のCSVファイルを data2 から探す。
    例:
      aim-2025-6-13-Fri_with_grape.csv
      aim-2025-06-13-Fri_with_grape.csv
      aim-2025-6-13-Fri.csv
      aim-2025-06-13-Fri.csv
    みたいなパターンを glob で拾う。
    """
    patterns = [
        f"aim-{d.year}-{d.month}-{d.day}-*.csv",      # 2025-6-13
        f"aim-{d:%Y-%m-%d}-*.csv",                    # 2025-06-13
    ]

    for pat in patterns:
        full_pat = os.path.join(DATA_DIR, pat)
        candidates = glob.glob(full_pat)
        if candidates:
            # 同じ日のファイルが複数ある場合はとりあえず最初の1つ
            return candidates[0]

    return None


def clean_int_col(df: pd.DataFrame, col: str) -> None:
    """カンマ・全角マイナス・空白などを除去して、nullable Int64 に変換"""
    if col not in df.columns:
        return

    s = (df[col].astype(str)
                 .str.replace("\u2212", "-", regex=False)   # 全角マイナス → 半角
                 .str.replace(",", "", regex=False)         # カンマ除去
                 .str.replace(r"\s+", "", regex=True))      # 空白除去

    # 先頭の符号付き整数だけを抜き出す（例: '-1028abc' → '-1028'）
    s = s.str.extract(r"([+-]?\d+)", expand=False)

    df[col] = pd.to_numeric(s, errors="coerce").astype("Int64")


def process_one_file(path: str):
    print(f"[INFO] 処理中: {os.path.basename(path)}")

    df = pd.read_csv(path, encoding="utf-8-sig")

    # 「平均」行を削除（台番が "平均" の行）
    if "台番" in df.columns:
        df = df[df["台番"] != "平均"].copy()

    # 整数として扱いたい列
    for c in ["台番", "差枚", "G数", "BB", "RB"]:
        clean_int_col(df, c)

    # 出力先ディレクトリ作成
    os.makedirs(OUTPUT_DIR, exist_ok=True)

    base = os.path.basename(path)
    name, ext = os.path.splitext(base)
    out_path = os.path.join(OUTPUT_DIR, name + "_clean" + ext)

    df.to_csv(out_path, index=False, encoding="utf-8-sig")
    print(f"[SAVE] {out_path}")


def main():
    start = datetime.strptime(START_DATE, "%Y-%m-%d")
    end   = datetime.strptime(END_DATE,   "%Y-%m-%d")

    for d in daterange(start, end):
        path = find_file_for_date(d)
        if path is None:
            print(f"[SKIP] {d:%Y-%m-%d} のファイルが見つかりません。")
            continue

        try:
            process_one_file(path)
        except Exception as e:
            print(f"[ERROR] {d:%Y-%m-%d} の処理でエラー: {e}")


if __name__ == "__main__":
    main()
```

```python
import os
import glob
import pandas as pd
from pathlib import Path

# 「平均」ラベルの行を集めて notebook/result/ave.csv に保存

def find_dir_up(start: Path, target_name: str) -> Path | None:
    """start から親方向へ target_name のディレクトリを探す"""
    cur = start.resolve()
    while True:
        cand = cur / target_name
        if cand.is_dir():
            return cand
        if cur.parent == cur:
            return None
        cur = cur.parent

def main():
    # 実行ディレクトリ（.py でも .ipynb でも対応）
    base_dir = Path(__file__).parent if "__file__" in globals() else Path.cwd()

    # 上方向にたどって data を特定（見つからなければ base_dir/data、無ければ base_dir）
    data_dir = find_dir_up(base_dir, "data") or (base_dir / "data")
    if not data_dir.is_dir():
        data_dir = base_dir  # 最後の保険

    # プロジェクトルートを推定（data のひとつ上）
    project_root = data_dir.parent if data_dir.name.lower() == "data" else base_dir

    # 出力先は notebook/result（なければ作る）。無ければ base_dir に落とす。
    output_dir = project_root / "notebook" / "result"
    try:
        output_dir.mkdir(parents=True, exist_ok=True)
    except Exception:
        output_dir = base_dir

    # CSV ファイル一覧（集計系は除外）
    files = sorted(
        f for f in glob.glob(str(data_dir / "*.csv"))
        if not Path(f).name.startswith(("ave", "mean_rows", "last_row_averages", "last_rows",
                                        "Filtered With Next Range", "aggregated"))
    )

    extracted = []
    for fp in files:
        # 文字コードを優先順で試す
        try:
            df = pd.read_csv(fp, encoding="utf-8-sig")
        except Exception:
            df = pd.read_csv(fp, encoding="cp932", errors="replace")
        if df.empty:
            continue

        # 先頭列の値が「平均」の行だけ抽出
        first_col = df.columns[0]
        mean_rows = df[df[first_col] == "平均"]
        if mean_rows.empty:
            continue

        # 元ファイル名を先頭列に付与
        mean_rows = mean_rows.copy()
        mean_rows.insert(0, "filename", Path(fp).name)
        extracted.append(mean_rows)

    if not extracted:
        print("「平均」ラベルの行が見つかりませんでした。")
        return

    result = pd.concat(extracted, ignore_index=True)
    print(result)

    out_fp = output_dir / "ave.csv"
    result.to_csv(out_fp, index=False, encoding="utf-8-sig")
    print(f"Saved mean rows to {out_fp}")

if __name__ == "__main__":
    main()
```

```python
from pathlib import Path
from datetime import datetime

# === 出力先 notebook/result を堅牢に解決 ===
def resolve_project_root() -> Path:
    """
    実行場所が:
      - プロジェクト直下
      - notebook フォルダ内
      - notebook/result フォルダ内
    のいずれでも、プロジェクト直下 Path を返す
    """
    try:
        here = Path(__file__).resolve()
        cwd = here.parent
    except NameError:
        # Notebook 実行などで __file__ が無い場合
        cwd = Path.cwd().resolve()

    # notebook/result から来た
    if cwd.name == "result" and cwd.parent.name == "notebook":
        return cwd.parent.parent
    # notebook から来た
    if cwd.name == "notebook":
        return cwd.parent
    # 直下に notebook がある＝プロジェクト直下
    if (cwd / "notebook").exists():
        return cwd
    # 念のため一つ上も見る
    if (cwd.parent / "notebook").exists():
        return cwd.parent

    # 最後のフォールバック
    return cwd

def save_dataframe_to_result(df, filename: str = None) -> Path:
    root = resolve_project_root()
    out_dir = root / "notebook" / "result"
    out_dir.mkdir(parents=True, exist_ok=True)

    # ファイル名が未指定ならタイムスタンプ付きにする
    if not filename:
        filename = f"rb_summary_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"

    out_path = out_dir / filename
    df.to_csv(out_path, index=False, encoding="utf-8-sig")
    print("Saved:", out_path)
    return out_path

# === ここで df_sorted を保存 ===
# すでに df_sorted がメモリにある前提
save_dataframe_to_result(df_sorted, "rb_summary.csv")
```
